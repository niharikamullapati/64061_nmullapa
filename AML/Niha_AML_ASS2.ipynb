**NEURAL NETWORKS**


The IMDB dataset


```python
from tensorflow.keras.datasets import imdb
(train2_data, train2_labels), (test2_data, test_labels) = imdb.load_data(num_words=10000)
```

```python
train2_data[0]
```

```python
train2_labels[0]
```

```python
#vectorizing the data
import numpy as np

def vectorize_sequences(sequences, dimension=10000):
    # Create an all-zero matrix of shape (len(sequences), dimension)
    res = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        res[i, sequence] = 1.  # set specific indices of results[i] to 1s
    return res

# Our vectorized training data
x_train2 = vectorize_sequences(train2_data)
# Our vectorized test data
x_test2 = vectorize_sequences(test2_data)
```

```python
x_train2[0]
```

```python
#vectorizing the labels
y_train = np.asarray(train2_labels).astype('float32')
y_test = np.asarray(test_labels).astype('float32')
```

**1. One and three hidden layers**

```python
from keras import models
from keras import layers

model_onehidden_layer = models.Sequential()
model_onehidden_layer.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model_onehidden_layer.add(layers.Dense(1, activation='sigmoid'))

model_onehidden_layer.compile(optimizer="rmsprop",
                               loss="binary_crossentropy",
                               metrics=["accuracy"])

x_val = x_train2[:10000]
partial_x_train = x_train2[10000:]

y_val = y_train[:10000]
partial_y_train = y_train[10000:]

history_one_hidden_layer = model_onehidden_layer.fit(partial_x_train,
                                                      partial_y_train,
                                                      epochs=20,
                                                      batch_size=512,
                                                      validation_data=(x_val, y_val))
import matplotlib.pyplot as plt
def plot_history(history):
    history_dict = history.history
    loss_values = history_dict["loss"]
    val_loss_values = history_dict["val_loss"]
    epochs = range(1, len(loss_values) + 1)

    plt.plot(epochs, loss_values, "bo", label="Training loss")
    plt.plot(epochs, val_loss_values, "b", label="Validation loss")
    plt.title("Training and validation loss")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()

    plt.clf()
    acc = history_dict["accuracy"]
    val_acc = history_dict["val_accuracy"]

    plt.plot(epochs, acc, "bo", label="Training acc")
    plt.plot(epochs, val_acc, "b", label="Validation acc")
    plt.title("Training and validation accuracy")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.show()


plot_history(history_one_hidden_layer)
model_three_hidden_layer = models.Sequential()
model_three_hidden_layer.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model_three_hidden_layer.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model_three_hidden_layer.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model_three_hidden_layer.add(layers.Dense(1, activation='sigmoid'))

model_three_hidden_layer.compile(optimizer="rmsprop",
                               loss="binary_crossentropy",
                               metrics=["accuracy"])

history_three_hidden_layer = model_three_hidden_layer.fit(partial_x_train,
                                                      partial_y_train,
                                                      epochs=20,
                                                      batch_size=512,
                                                      validation_data=(x_val, y_val))
plot_history(history_three_hidden_layer)


```

**3. Using the MSE loss function instead of binary_crossentropy.**

```python
from keras import models
from keras import layers

model_mse_loss = models.Sequential()
model_mse_loss.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model_mse_loss.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model_mse_loss.add(layers.Dense(1, activation='sigmoid'))

model_mse_loss.compile(optimizer="rmsprop",
                               loss="mse",
                               metrics=["accuracy"])

history_mse_loss = model_mse_loss.fit(partial_x_train,
                                      partial_y_train,
                                      epochs=20,
                                      batch_size=512,
                                      validation_data=(x_val, y_val))
```

```python
plot_history(history_mse_loss)
```

**4. Using the tanh activation instead of relu.**

```python
model_tanh_activation = models.Sequential()
model_tanh_activation.add(layers.Dense(32, activation='tanh', input_shape=(10000,)))
model_tanh_activation.add(layers.Dense(32, activation='tanh', input_shape=(10000,)))
model_tanh_activation.add(layers.Dense(1, activation='sigmoid'))

model_tanh_activation.compile(optimizer="rmsprop",
                               loss="mse",
                               metrics=["accuracy"])

history_tanh_activation = model_tanh_activation.fit(partial_x_train,
                                      partial_y_train,
                                      epochs=20,
                                      batch_size=512,
                                      validation_data=(x_val, y_val))
```

```python
plot_history(history_tanh_activation)
```

**5. Using Regularization and Dropout.**

```python
from keras import regularizers
model_regularization_dropout = models.Sequential([
    layers.Dense(32, activation="tanh", kernel_regularizer=regularizers.l2(0.001)),
    layers.Dropout(0.5),
    layers.Dense(32, activation="tanh", kernel_regularizer=regularizers.l2(0.001)),
    layers.Dropout(0.5),
    layers.Dense(1, activation="sigmoid")])
model_regularization_dropout.compile(optimizer="rmsprop", loss="binary_crossentropy", metrics=["accuracy"])
history_regularization_dropout = model_regularization_dropout.fit(partial_x_train,
                                                                  partial_y_train,
                                                                  epochs=20,
                                                                  batch_size=512,
                                                                  validation_data=(x_val, y_val))
```

```python
plot_history(history_regularization_dropout )
```

**SUMMARY :**

For this exercise, various configurations of neural networks were analysiert using the IMDB dataset, with adjustments made to the number of hidden layers and units as well as to the activation and loss functions. Even with the use of regularization methods such as L2 adjustment and Dropout, the model showed signs of overfitting after several epochs. This became clear as the validation accuracy leveled off and the validation loss rose.
The graphs reveal that
Fewer Hidden Units (16 units): By the conclusion of training, the model achieved a training accuracy nearing 100%.
However, the validation accuracy reached a plateau at about 87%, while the validation loss increased sharply in the final stages, signaling overfitting.  This indicates that the model performed well on the training data but struggled to generalize to new data..
More Hidden Units (64 units): This model reached a training accuracy of 100% as well
Utilizing MSE Loss Function Rather than Binary Cross-Entropy: The model's performance showed slight variations in loss values as a result of the modification in the loss function.
The validation accuracy was comparable to that of the models employing binary cross-entropy, achieving approximately 87-88%. The validation loss stayed within acceptable limits, albeit a bit higher than that of the binary cross-entropy model.
Utilizing Tanh Activation Rather than Relu: The model utilizing the tanh activation function attained a high level of accuracy, comparable to that of the relu-based models.
The validation accuracy and loss were similar, indicating that altering the activation function had minimal impact on performance in this case.
In general, models with fewer hidden units as well as those with more hidden units yielded comparable results, demonstrating that increasing the number of hidden units did not lead to a significant enhancement.
. Nevertheless, every model exhibited overfitting.  The model that used MSE as the loss function performed somewhat worse than the one that used binary cross-entropy. The model with tanh activation had performance comparable to that of the relu-based model.
