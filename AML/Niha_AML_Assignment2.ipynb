```python
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

# Generate an artificial dataset with 3 classes and centers
centers = [[2, 4], [6, 4], [1, 6]]
n_classes = len(centers)
data, labels = make_blobs(n_samples=150,
                          centers=np.array(centers),
                          random_state=1)

# Spliting the dataset into 80% training set and 20% testing set
train_data, test_data, train_labels, test_labels = train_test_split(data, labels,
                                                                    train_size=0.7,
                                                                    test_size=0.3,
                                                                    random_state=13)

# Performing  KNN analysis
knn1 = KNeighborsClassifier(n_neighbors=5)  # using 5 nearest neighbors
knn1.fit(train_data, train_labels)

#  print Predictions and Accuracy for the training set
print("Predictions from the classifier:")
train_predictions = knn1.predict(train_data)
print(train_predictions)
print(train_labels)
print("Accuracy on training set:", accuracy_score(train_predictions, train_labels))

# Print predictions and accuracy for the testing set
test_predictions = knn1.predict(test_data)
print("Predictions on test set:")
print(test_predictions)
print("Target values:")
print(test_labels)
print("Accuracy on test set:", accuracy_score(test_predictions, test_labels))

# Plot the results
plt.figure(figsize=(14, 7))

# Plot the training data with true labels
plt.subplot(1, 2, 1)
plt.scatter(train_data[:, 0], train_data[:, 1], c=train_labels, cmap='viridis', edgecolor='k', s=100, marker='o')
plt.title('Training Data with True Labels')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar()

# Plot the test data with predicted labels
plt.subplot(1, 2, 2)
plt.scatter(test_data[:, 0], test_data[:, 1], c=test_predictions, cmap='viridis', edgecolor='k', s=100, marker='s')
plt.title('Test Data with Predicted Labels')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar()

plt.tight_layout()
plt.show()

# Plot the data points with different colors for each class
plt.figure(figsize=(8, 6))
for i in range(n_classes):
    plt.scatter(data[labels == i][:, 0], data[labels == i][:, 1], label=f'Class {i}')
plt.title('Simulated Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
